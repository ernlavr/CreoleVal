```python
with open('ht/wikiann-ht.bio', 'r', encoding='utf8') as f:
    txt = []
    for token in f:
        txt.append(token)              #getting the file
txt
```

```python
sents = {}
k = 1
for tok in txt:
    if tok != '\n':
        sents.setdefault(k, []).append(tok)
    elif tok == '\n':
        k += 1                  #creating a dict of all the individual instances
len(sents)
```

```python
unique_sents = {}
for key, value in sents.items():
    value = '+'.join(value)
    if value not in unique_sents:
        unique_sents[value] = 1
    elif value in unique_sents:
        unique_sents[value] += 1    #creating a dict with the frequency of each instance to weed out repeats
len(unique_sents)
```

```python
delkeys = []
for key, value in sents.items():
    value = '+'.join(value)
    if unique_sents[value] > 1:
        delkeys.append(key)        #list with keys of all the repeated instances
len(delkeys)
```

```python
for key in delkeys:
    if key in sents:
        del sents[key]       #removing all the instances that are repeated
len(sents)
```

```python
duplicates = {}
n = 0
for key, value in unique_sents.items():
    if value > 1:
        key = key.split('+')
        n += 1
        duplicates[n] = key      #creating another dict with all the repeated instances
len(duplicates)
```

```python
nkey = 100000
for key, value in duplicates.items():
    nkey += 1
    sents[nkey] = value        #adding the repeated instances back into the sents dict, but only once
len(sents)
```

```python
newsents = {}
ke = 0
for key, value in sents.items():
    ke +=1
    newsents[ke] = value    #sorting the keys in a new dict to split the text
len(newsents)
```

```python
nr = 0
lengths = {}
for key, value in newsents.items():
    nr += 1
    lengths[nr] = len(value)
```
